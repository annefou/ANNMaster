```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv('../data/met_data_ANN.csv')
```

```python
df=df.set_index('Dato')
```

```python
df.index=  pd.to_datetime(df.index, format='%Y-%m-%d %H:%M:%S.%f')
```

```python
df
```

```python
fig = plt.figure(figsize=(15,7))
ax1 = fig.add_subplot(111)
ax1.plot(df.index,df['T_air'], color='r')   
ax1.set_xticks(df.index)
ax1.set_xticklabels(df.index, rotation=45, fontsize=10 )
ax2 = ax1.twinx()
ax2.plot(df.index,df['RH'])
# Now add the legend with some customizations.
legend= fig.legend(loc='upper center', shadow=True)
```

```python
fig, ax = plt.subplots(nrows=5,
                         ncols=1,
                         sharex=True)
fig.set_figheight(12)
fig.set_figwidth(17)
df['LW_down'].plot(ax=ax[0], linewidth=2, color='r', linestyle='solid')
df['SW_down'].plot(ax=ax[1], linewidth=2, color='b', linestyle='solid')
df['WS'].plot(ax=ax[2], linewidth=2, color='g', linestyle='solid')
df['T_air'].plot(ax=ax[3], linewidth=2, color='m', linestyle='solid')
df['RH'].plot(ax=ax[4], linewidth=2, color='b', linestyle='solid')


# Now add the legend with some customizations.
legend= fig.legend(loc='upper center', shadow=True)
```
